```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取第三个数据文件
def read_csv_file(path):
    # '../pop/APG-SMOEA_pop_gen_100.csv'
    df = pd.read_csv(path)
    
    # 将return的负数转换为正数
    df['return'] = df['return'].apply(abs)

    returns = df['return']
    risks = df['risk']

    return risks, returns
```
